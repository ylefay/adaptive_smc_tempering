```python
import pickle
import os
import matplotlib.pyplot as plt
import numpy as np
import jax.numpy as jnp
import pandas as pd

PKLs = []
PKL_titles = []
DIR = "./output/"
for file in os.listdir(DIR):
    if file.endswith(".pkl"):
        PKLs.append(pickle.load(open(DIR+file, "rb")))
        PKL_titles.append(str(file))
n = len(PKLs)
plt.rcParams['text.usetex'] = False
```

```python
concat_temp = jnp.concat([PKLs[idx]['res'][6] for idx in range(1, n)])
concat_lg_norm = jnp.concat([PKLs[idx]['res'][8] for idx in range(1, n)])

res = {'res': (None, None, None, None, None, None, concat_temp, None, concat_lg_norm), 'config': PKLs[1]['config']}
with open(DIR+"merge.pkl", "wb") as handle:
    pickle.dump(res, handle, protocol=pickle.HIGHEST_PROTOCOL)
```
