```python
import pickle
import os
import matplotlib.pyplot as plt
import numpy as np
import jax.numpy as jnp
import pandas as pd

%env XLA_PYTHON_CLIENT_PREALLOCATE=False


PKLs = []
PKL_titles = []
DIR = "./output_28_dim10_mean0_test/"
for file in os.listdir(DIR):
    if file.endswith(".pkl"):
        loaded = pickle.load(open(DIR+file, "rb"))
        loaded =  {'res': (None, None, None, None, None, loaded['res'][5], loaded['res'][6], None, loaded['res'][8]), 'config':loaded['config']}
        PKLs.append(loaded)
        PKL_titles.append(str(file))
n = len(PKLs)
plt.rcParams['text.usetex'] = False
```

```python
idx_proposal = dict()
for i in range(n):
    proposal = PKLs[i]['config']['proposal']
    idx_proposal[proposal] = idx_proposal.get(proposal, []) + ([i])

for proposal in idx_proposal:
    concat_mh = jnp.concat([PKLs[idx]['res'][3] for idx in idx_proposal[proposal]])
    concat_temp = jnp.concat([PKLs[idx]['res'][6] for idx in idx_proposal[proposal]])
    concat_lg_norm = jnp.concat([PKLs[idx]['res'][8] for idx in idx_proposal[proposal]])
    res = {'res': (None, None, None, concat_mh, None, None, concat_temp, None, concat_lg_norm), 'config': PKLs[idx_proposal[proposal][0]]['config']}
    with open(DIR+proposal+"_merge.pkl", "wb") as handle:
        pickle.dump(res, handle, protocol=pickle.HIGHEST_PROTOCOL)
```
