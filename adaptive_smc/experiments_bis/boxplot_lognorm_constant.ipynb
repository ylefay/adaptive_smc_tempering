```python
import pickle
import os
import matplotlib.pyplot as plt
import numpy as np
import jax.numpy as jnp
import pandas as pd

PKLs = []
PKL_titles = []
DIR = "./gaussians/output/"
for file in os.listdir(DIR):
    if file.endswith(".pkl"):
        PKLs.append(pickle.load(open(DIR+file, "rb")))
        PKL_titles.append(str(file))
n = len(PKLs)
plt.rcParams['text.usetex'] = True
```

```python
list_idx_rwmh = []
list_idx_ar = []
for i in range(n):
    if 'rw' in PKLs[i]['config']['proposal']:
        list_idx_rwmh.append(i)
    else:
        list_idx_ar.append(i)
        
n_ar = len(list_idx_ar)
n_rw = len(list_idx_rwmh)
```

```python
PKLs[list_idx_ar[0]]['res'][8].shape
```

```python
max_size = PKLs[i]['res'][6].shape[1]
num_para = PKLs[i]['res'][0].shape[0]
max_temp_idx = max([jnp.min(jnp.argwhere(PKLs[i]['res'][6]==1), axis=0)[1] for i in range(n)])
log_normalization_constants_ar = np.array([PKLs[list_idx_ar[i]]['res'][8] for i in range(n_ar)])
log_normalization_constants_ar = log_normalization_constants_ar.reshape((n_ar * num_para, max_size+1))
log_normalization_constants_rwmh = np.array([PKLs[list_idx_rwmh[i]]['res'][8] for i in range(len(list_idx_rwmh))])
log_normalization_constants_rwmh = log_normalization_constants_rwmh.reshape((n_rw * num_para, max_size+1))
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
df1 = pd.DataFrame(log_normalization_constants_rwmh.reshape((n_rw * num_para, max_size+1))[...,1:max_temp_idx], columns=list(range(1,max_temp_idx))).assign(Type='RW')
df2 = pd.DataFrame(log_normalization_constants_ar.reshape((n_ar * num_para, max_size+1))[...,1:max_temp_idx], columns=list(range(1,max_temp_idx))).assign(Type='AR') # we get rid of the first iteration t=0, no difference can be observed there, 1:max_tmep_idx

# Sample DataFrames
df1 = df1

df2 = df2

# Combine the DataFrames
combined_df = pd.concat([df1, df2])

# Melt the combined DataFrame to long-form
melted_df = combined_df.melt(id_vars=['Type'], var_name='Variable', value_name='Value')

# Create the boxplot
plt.figure(figsize=(10, 6))
sns.boxplot(x='Variable', y='Value', hue='Type', data=melted_df)

# Add titles and labels
plt.title('Boxplot of log normalization constant at each iteration')
plt.xlabel('Iteration')
plt.ylabel(r'log norm. constant ($\log N^{-1} \sum w_t^n)')

# Show the plot
plt.savefig(DIR+"boxplot.png")
```
